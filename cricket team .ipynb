```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Generate some sample data for the bar chart
x = np.arange(4)
team1_wickets = [3, 6, 10, 10]
team2_wickets = [7, 10, 4, 3]

# Define colors for each bar
colors = ['skyblue']

# Create the bar chart
plt.bar(x, team1_wickets, color=colors, label='Team 1 Wickets')
plt.bar(x, team2_wickets, color='red', label='Team 2 Wickets', alpha=0.5)  # Overlaying bar chart for team 2 wickets

# Add labels and title
plt.xlabel('Matches')
plt.ylabel('Wickets')
plt.title('Premium Vector Graph')
plt.legend()

# Define the directory path
directory = r'C:\path\to\save'

# Create the directory if it doesn't exist
if not os.path.exists(directory):
    os.makedirs(directory)

# Save the plot as an image file
plt.savefig(os.path.join(directory, 'premium_vector_graph.png'))

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Sample data
sizes = np.random.randint(1, 10, size=4)
labels = ['Sri Lanka', 'India', 'New Zealand', 'Australia']

# Define colors for each section of the donut chart
colors = ['purple', 'gray', 'orange', 'black']

# Create the donut chart
plt.pie(sizes, labels=labels, colors=colors, wedgeprops=dict(width=0.3), autopct='%1.1f%%')

# Add title
plt.title('Donut Chart')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Define the directory path
directory = r'C:\path\to\save'

# Create the directory if it doesn't exist
if not os.path.exists(directory):
    os.makedirs(directory)

# Save the plot as an image file
plt.savefig(os.path.join(directory, 'donut_chart.png'))

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

# Sample data
x = np.arange(4)  # Assuming you have 4 dates
y1 = np.random.randint(1, 10, size=4)  # Generating random y-values for area 1
y2 = np.random.randint(1, 10, size=4)  # Generating random y-values for area 2

# Sample data
Umpire1 = ['JD Cloete', 'RJ Bailey', 'JM Williams', 'Ahmed Shah Durrani']
Umpire2 = ['NJ Llong', 'CB Gaffaney', 'RR Wimalasiri', 'C Howard']

# Create the area graph
plt.fill_between(x, y1, color='purple', alpha=0.5, label='Area 1')
plt.fill_between(x, y2, color='black', alpha=0.5, label='Area 2')

# Add labels and title
plt.xlabel('Umpire 1')
plt.ylabel('Umpire 2')
plt.title('Premium Vector Graph')
plt.xticks(x, Umpire1)  # Set the x-axis ticks as Umpire 1 names

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend
plt.legend()

# Define the directory path
directory = r'C:\path\to\save'

# Create the directory if it doesn't exist
if not os.path.exists(directory):
    os.makedirs(directory)

# Save the plot as an image file
plt.savefig(os.path.join(directory, 'premium_vector_graph.png'))

# Show the plot
plt.show()
```
